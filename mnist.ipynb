```python
from sklearn.datasets import   load_digits
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import  TSNE
from sklearn.decomposition import PCA
```

```python
digits = load_digits() # 8x8 images of digits 0-9 
X = digits.data 
y = digits.target
```

```python
# dimension 
print(X.shape)
```

```python
X_standard = X/255.0
```

```python
# Target is the label of the data
np.unique(y)
```

```python
# visualize the data 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standard)
```

```python
# plot the data
plt.figure(figsize=(10,10))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# apply t-SNE
# perplexity is the number of nearest neighbors that is used in other to compute the distances
# range [5-50]
# max_iter is the maximum number of iterations to run
tsne = TSNE(n_components=2,perplexity=20,max_iter=3000)
X_tsne = tsne.fit_transform(X_standard)
print(tsne.kl_divergence_) 
# kl_divergence_ is the Kullback-Leibler divergence of the model 
# higher the value, the worse the model is
# lower the value, the better the model is
```

```python
# plot the data 
plt.figure(figsize=(10,10))
scatter =plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='viridis',s=1)
plt.legend(*scatter.legend_elements(), title="Digits")
plt.colorbar()
plt.show()
```
